```python
import numpy as np
import scipy.io
import os, glob

def load_ppg_data(folder_path, sample_len=256, num_files=None):
    """
    Load segmented PPG data and labels from .mat files in a folder.

    Args:
        folder_path (str): Path to directory containing .mat files.
        sample_len (int): Length of each segment.
        num_files (int or None): Number of files to load (default: all).
        
    Returns:
        X (np.ndarray): PPG segments.
        y (np.ndarray): Corresponding labels.
    """
    X, y = [], []
    mat_files = glob.glob(os.path.join(folder_path, "*.mat"))
    if num_files is not None:
        mat_files = mat_files[:num_files]

    for file in mat_files:
        data = scipy.io.loadmat(file)
        ppg = data['ppg_signals'].flatten()
        labels = data['sleep_stages'].flatten()

        for i in range(0, len(ppg) - sample_len, sample_len):
            segment = ppg[i:i+sample_len]
            label_idx = i // sample_len
            if label_idx < len(labels):
                X.append(segment)
                y.append(labels[label_idx])

    return np.array(X), np.array(y)
```

```python
def quantize_signal(signals, num_bits=8):
    """
    Quantize signal into integer tokens.

    Args:
        signals (np.ndarray): Input signal array.
        num_bits (int): Number of bits for quantization.
        
    Returns:
        quantized (np.ndarray): Quantized signal.
        scale (float): Scaling factor used.
        min_val (float): Minimum of original signal.
    """
    min_val, max_val = np.min(signals), np.max(signals)
    qmin, qmax = 0, 2**num_bits - 1
    scale = (max_val - min_val) / (qmax - qmin)
    quantized = ((signals - min_val) / scale).astype(np.uint8)
    return quantized, scale, min_val
```

```python
import random

def jitter(x, sigma=0.03):
    return x + np.random.normal(0, sigma, size=x.shape)

def scaling(x, sigma=0.1):
    return x * np.random.normal(1.0, sigma)

def magnitude_warp(x, sigma=0.2):
    return x * (1 + sigma * np.sin(np.linspace(0, 3 * np.pi, len(x))))

def slice_random(x, size=256):
    start = np.random.randint(0, len(x) - size)
    return x[start:start+size]

def augment_batch(X):
    augmented = []
    for x in X:
        method = random.choice([jitter, scaling, magnitude_warp])
        x_aug = method(x)
        if len(x_aug) > 256:
            x_aug = slice_random(x_aug)
        augmented.append(x_aug)
    return np.array(augmented)
```

```python
import tensorflow as tf

def create_tf_dataset(X, y, batch_size=32, shuffle=True):
    ds = tf.data.Dataset.from_tensor_slices((X, y))
    if shuffle:
        ds = ds.shuffle(buffer_size=1000)
    ds = ds.batch(batch_size).prefetch(tf.data.AUTOTUNE)
    return ds
```

```python
from tensorflow.keras import layers, models

def create_tiny_transformer(input_length=256, vocab_size=256, d_model=64, num_heads=4, ff_dim=128, num_layers=2, num_classes=5):
    inputs = layers.Input(shape=(input_length,), dtype=tf.int32)

    # Token Embedding
    x = layers.Embedding(input_dim=vocab_size, output_dim=d_model)(inputs)

    # Positional Embedding
    pos = tf.range(start=0, limit=input_length, delta=1)
    pos_emb = layers.Embedding(input_length, d_model)(pos)
    x = x + pos_emb

    # Transformer Encoder Layers
    for _ in range(num_layers):
        attn_out = layers.MultiHeadAttention(num_heads=num_heads, key_dim=d_model)(x, x)
        x = layers.Add()([x, attn_out])
        x = layers.LayerNormalization()(x)

        ffn = layers.Dense(ff_dim, activation='relu')(x)
        ffn = layers.Dense(d_model)(ffn)
        x = layers.Add()([x, ffn])
        x = layers.LayerNormalization()(x)

    x = layers.GlobalAveragePooling1D()(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)

    return models.Model(inputs, outputs)
```

```python
import tensorflow.keras.backend as K

def focal_loss(gamma=2.0, alpha=0.25):
    def loss(y_true, y_pred):
        y_true = tf.squeeze(y_true, axis=-1)  # (batch,)
        y_true = tf.one_hot(tf.cast(y_true, tf.int32), depth=tf.shape(y_pred)[-1])  # (batch, num_classes)

        # Clip predictions to prevent log(0) error
        epsilon = K.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)

        cross_entropy = -y_true * tf.math.log(y_pred)  # (batch, num_classes)
        weights = alpha * tf.pow(1 - y_pred, gamma)  # (batch, num_classes)

        loss = weights * cross_entropy  # (batch, num_classes)
        return tf.reduce_mean(tf.reduce_sum(loss, axis=-1))  # mean over batch
    return loss
```

```python
def compile_and_train(model, train_ds, val_ds, epochs=20):
    model.compile(
        optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
        loss=focal_loss(gamma=2.5, alpha=0.5),
        metrics=['accuracy']
    )
    history = model.fit(train_ds, validation_data=val_ds, epochs=epochs)
    return history
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (classification_report, confusion_matrix, 
                             accuracy_score, f1_score, precision_score, recall_score,
                             cohen_kappa_score, matthews_corrcoef, roc_auc_score)

def evaluate_full(model, X_test, y_test, num_classes=5, is_softmax=True):
    y_probs = model.predict(X_test)
    y_pred = y_probs.argmax(axis=1) if is_softmax else (y_probs > 0.5).astype(int)

    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='weighted')
    prec = precision_score(y_test, y_pred, average='weighted')
    rec = recall_score(y_test, y_pred, average='weighted')
    kappa = cohen_kappa_score(y_test, y_pred)
    mcc = matthews_corrcoef(y_test, y_pred)

    print("\nüìä Classification Report:\n", classification_report(y_test, y_pred))
    print(f"‚úÖ Accuracy: {acc:.4f}")
    print(f"‚úÖ F1-score: {f1:.4f}")
    print(f"‚úÖ Precision: {prec:.4f}")
    print(f"‚úÖ Recall: {rec:.4f}")
    print(f"‚úÖ Cohen‚Äôs Kappa: {kappa:.4f}")
    print(f"‚úÖ MCC: {mcc:.4f}")

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("True")
    plt.show()

    try:
        y_test_onehot = tf.keras.utils.to_categorical(y_test, num_classes)
        auc = roc_auc_score(y_test_onehot, y_probs, multi_class="ovr", average="macro")
        print(f"‚úÖ ROC-AUC (macro avg): {auc:.4f}")
    except:
        print("‚ö†Ô∏è ROC-AUC skipped (shape mismatch or binary probs missing)")
```
