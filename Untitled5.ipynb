```python
import numpy as np
import os, glob, scipy.io, time
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, cohen_kappa_score
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, SeparableConv1D, MaxPooling1D, GlobalAveragePooling1D,
                                     Flatten, Dense, Dropout, Multiply, Add, LayerNormalization,
                                     Lambda, GaussianNoise)
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau

tf.keras.backend.clear_session()

# ✅ Ensure GPU Usage
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"✅ GPU detected: {gpus[0]}")
    tf.config.experimental.set_memory_growth(gpus[0], True)
else:
    print("⚠️ No GPU found. Training might be slow.")

# ⏱️ Start Timing
start_time = time.time()

# 📂 Load PPG Data
folder_path = r"C:\Users\DST\Desktop\micro_AK\mat files"
mat_files = sorted(glob.glob(os.path.join(folder_path, "*.mat")))[:50]

X_list, y_list = [], []
for file in mat_files:
    mat = scipy.io.loadmat(file)
    X_list.append(mat['ppg_signals'])
    y_list.append(mat['sleep_stages'].flatten())

X = np.vstack(X_list)
y = np.concatenate(y_list)

# ✅ Data Augmentation Functions
def jitter(x, sigma=0.01):
    return x + np.random.normal(loc=0, scale=sigma, size=x.shape)

def scaling(x, sigma=0.1):
    return x * np.random.normal(loc=1.0, scale=sigma, size=x.shape)

def magnitude_warp(x, sigma=0.2):
    x = x.reshape(-1)
    warp = np.sin(np.linspace(0, np.pi, len(x))) * np.random.normal(1, sigma)
    return x * warp

# 🔁 Class-wise Augmentation (focus on underrepresented)
np.random.seed(42)
augmented = []
target_distribution = np.bincount(y)
minority_classes = np.where(target_distribution < np.median(target_distribution))[0]

for signal, label in zip(X, y):
    if label in minority_classes:
        if np.random.rand() < 0.5:
            signal = jitter(signal)
        if np.random.rand() < 0.5:
            signal = magnitude_warp(signal)
        if np.random.rand() < 0.5:
            signal = scaling(signal.reshape(1, -1)).flatten()
    augmented.append(signal)

X = np.array(augmented)

# 🔄 Standardization
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 🧩 Handle Class Imbalance
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# 🧠 Class Weights
class_weights = compute_class_weight("balanced", classes=np.unique(y_res), y=y_res)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# 🔁 Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, stratify=y_res, random_state=42
)

# 🔷 Preprocessing
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=5)
y_test_cat = tf.keras.utils.to_categorical(y_test, num_classes=5)

# ⚠️ Focal Loss Function
def focal_loss(gamma=1.5, alpha=0.25):
    def loss_fn(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.math.pow(1 - y_pred, gamma)
        return tf.reduce_mean(weight * cross_entropy)
    return loss_fn

# 🚀 Updated Lightweight Hybrid CNN-RVFL Model with Attention
def build_hybrid_cnn_rvfl(input_shape):
    inputs = Input(shape=input_shape)
    x = GaussianNoise(0.05)(inputs)
    
    # SeparableConv blocks with strides to reduce dimension instead of MaxPooling for lightweight efficiency
    x = SeparableConv1D(32, 5, strides=2, padding='same', activation='relu')(x)
    x = SeparableConv1D(64, 3, strides=2, padding='same', activation='relu')(x)
    x = SeparableConv1D(128, 3, strides=2, padding='same', activation='relu')(x)
    x = Dropout(0.2)(x)  # SpatialDropout1D is not imported, use Dropout
    
    # Residual Connection with stride=2 Conv1D to match dims
    res = tf.keras.layers.Conv1D(128, 1, strides=8, padding='same')(inputs)
    
    # SE Attention
    se = GlobalAveragePooling1D()(x)
    se = Dense(64, activation='relu')(se)
    se = Dense(128, activation='sigmoid')(se)
    se = Lambda(lambda s: tf.expand_dims(s, 1))(se)
    x = Multiply()([x, se])
    
    # Add Residual and Normalize
    x = Add()([x, res])
    x = LayerNormalization()(x)
    
    # RVFL-like Head
    x = Flatten()(x)
    x = Dense(128, activation='relu')(x)
    x = Dropout(0.3)(x)
    x = Dense(64, activation='relu')(x)
    x = Dropout(0.2)(x)
    outputs = Dense(5, activation='softmax')(x)
    
    model = Model(inputs, outputs)
    return model

# ⚙️ Compile & Train
model = build_hybrid_cnn_rvfl((X_train.shape[1], 1))
model.compile(optimizer=Adam(1e-4), loss=focal_loss(), metrics=['accuracy'])

model.summary()

callbacks = [
    EarlyStopping(patience=8, restore_best_weights=True),
    ModelCheckpoint("best_model.keras", save_best_only=True),
    ReduceLROnPlateau(patience=3, factor=0.5)
]

history = model.fit(X_train, y_train_cat, validation_data=(X_test, y_test_cat),
                    epochs=70, batch_size=32, class_weight=class_weight_dict,
                    callbacks=callbacks, verbose=1)

# 📉 Training Curves
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='train')
plt.plot(history.history['val_accuracy'], label='val')
plt.title("Accuracy")
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='train')
plt.plot(history.history['val_loss'], label='val')
plt.title("Loss")
plt.legend()
plt.tight_layout()
plt.show()

# 📊 Evaluation
y_pred = model.predict(X_test)
y_pred_cls = np.argmax(y_pred, axis=1)
y_true_cls = np.argmax(y_test_cat, axis=1)

print("\nClassification Report:")
print(classification_report(y_true_cls, y_pred_cls))

cm = confusion_matrix(y_true_cls, y_pred_cls)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.show()

kappa = cohen_kappa_score(y_true_cls, y_pred_cls)
print(f"\n🧠 Cohen Kappa Score: {kappa:.4f}")

# 💾 Save for Deployment
model.save("cnn_rvfl_final_model.keras")

# ✅ Export for TFLite (Edge Devices)
converter = tf.lite.TFLiteConverter.from_keras_model(model)
tflite_model = converter.convert()
with open("cnn_rvfl_model.tflite", "wb") as f:
    f.write(tflite_model)

# ⏱️ Total Training Time
end_time = time.time()
print(f"\n⏱️ Total training time: {(end_time - start_time)/60:.2f} minutes")
```

```python
import numpy as np
import os, glob, scipy.io, time
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, cohen_kappa_score, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import RidgeClassifierCV

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, SeparableConv1D, GlobalAveragePooling1D, BatchNormalization,
                                     Dropout, Multiply, Add, LayerNormalization, Dense, Lambda,
                                     Flatten, GaussianNoise, Concatenate)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import initializers

tf.keras.backend.clear_session()

# Ensure GPU
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"✅ GPU detected: {gpus[0]}")
    tf.config.experimental.set_memory_growth(gpus[0], True)
else:
    print("⚠️ No GPU found. Training might be slow.")

start_time = time.time()

# Load Data
folder_path = r"C:\Users\DST\Desktop\micro_AK\mat files"
mat_files = sorted(glob.glob(os.path.join(folder_path, "*.mat")))[:50]

X_list, y_list = [], []
for file in mat_files:
    mat = scipy.io.loadmat(file)
    X_list.append(mat['ppg_signals'])
    y_list.append(mat['sleep_stages'].flatten())

X = np.vstack(X_list)
y = np.concatenate(y_list)

# Augmentation
def jitter(x, sigma=0.01): return x + np.random.normal(0, sigma, x.shape)
def scaling(x, sigma=0.1): return x * np.random.normal(1.0, sigma, x.shape)
def magnitude_warp(x, sigma=0.2):
    x = x.reshape(-1)
    warp = np.sin(np.linspace(0, np.pi, len(x))) * np.random.normal(1, sigma)
    return x * warp

np.random.seed(42)
augmented = []
target_distribution = np.bincount(y)
minority_classes = np.where(target_distribution < np.median(target_distribution))[0]

for signal, label in zip(X, y):
    if label in minority_classes:
        if np.random.rand() < 0.5:
            signal = jitter(signal)
        if np.random.rand() < 0.5:
            signal = magnitude_warp(signal)
        if np.random.rand() < 0.5:
            signal = scaling(signal.reshape(1, -1)).flatten()
    augmented.append(signal)

X = np.array(augmented)

# Normalize
scaler = StandardScaler()
X = scaler.fit_transform(X)

# SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Class weights
class_weights = compute_class_weight("balanced", classes=np.unique(y_res), y=y_res)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, stratify=y_res, random_state=42
)
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
num_classes = 5
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_test_cat = tf.keras.utils.to_categorical(y_test, num_classes=num_classes)

# Focal Loss
def focal_loss(gamma=1.5, alpha=0.25):
    def loss_fn(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.math.pow(1 - y_pred, gamma)
        return tf.reduce_mean(weight * cross_entropy)
    return loss_fn

# Fixed Random Layer for RVFL
class FixedRandomDense(tf.keras.layers.Layer):
    def __init__(self, units, activation='relu', **kwargs):
        super(FixedRandomDense, self).__init__(**kwargs)
        self.units = units
        self.activation = tf.keras.activations.get(activation)
        self.activation_str = activation

    def build(self, input_shape):
        self.w = self.add_weight(
            shape=(int(input_shape[-1]), self.units),
            initializer=initializers.RandomNormal(stddev=0.05),
            trainable=False,
            name='fixed_w'
        )
        self.b = self.add_weight(
            shape=(self.units,),
            initializer='zeros',
            trainable=False,
            name='fixed_b'
        )

    def call(self, inputs):
        return self.activation(tf.matmul(inputs, self.w) + self.b)

    def get_config(self):
        config = super(FixedRandomDense, self).get_config()
        config.update({
            "units": self.units,
            "activation": self.activation_str
        })
        return config

# Build CNN-RVFL Hybrid Model
def build_cnn_rvfl(input_shape, rvfl_units=256, final_units=num_classes):
    inputs = Input(shape=input_shape)
    x = GaussianNoise(0.05)(inputs)

    # CNN Feature Extractor
    x = SeparableConv1D(32, 5, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = SeparableConv1D(64, 3, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = SeparableConv1D(128, 3, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.25)(x)
    x = BatchNormalization()(x)

    # Residual connection
    res = tf.keras.layers.Conv1D(128, 1, strides=8, padding='same')(inputs)
    if x.shape[1] != res.shape[1]:
        res = tf.keras.layers.Resizing(x.shape[1], x.shape[2])(res)
    x = Add()([x, res])
    x = LayerNormalization()(x)

    # SE Attention
    se = GlobalAveragePooling1D()(x)
    se = Dense(64, activation='relu')(se)
    se = Dense(128, activation='sigmoid')(se)
    se = Lambda(lambda s: tf.expand_dims(s, 1))(se)
    x = Multiply()([x, se])
    x = GlobalAveragePooling1D()(x)

    # RVFL Branch
    x_flat = Flatten()(inputs)
    x_rvfl = FixedRandomDense(rvfl_units, activation='relu')(x_flat)

    # Merge
    concat = Concatenate()([x, x_rvfl])
    outputs = Dense(final_units, activation='softmax')(concat)

    model = Model(inputs, outputs)
    return model

# Build and compile model
model = build_cnn_rvfl((X_train.shape[1], 1), rvfl_units=256, final_units=num_classes)
model.compile(optimizer=Adam(1e-4), loss=focal_loss(), metrics=['accuracy'])
model.summary()

# Callbacks
callbacks = [
    EarlyStopping(patience=8, restore_best_weights=True),
    ModelCheckpoint("best_model.keras", save_best_only=True),
    ReduceLROnPlateau(patience=3, factor=0.5)
]

# Train model
history = model.fit(X_train, y_train_cat, validation_data=(X_test, y_test_cat),
                    epochs=100, batch_size=32, class_weight=class_weight_dict,
                    callbacks=callbacks, verbose=1)

# Extract features before classifier
intermediate_model = Model(inputs=model.input, outputs=model.layers[-2].output)
X_train_feats = intermediate_model.predict(X_train)

# Ridge classifier
ridge = RidgeClassifierCV(alphas=np.logspace(-6, 6, 13), scoring='accuracy', cv=5)
ridge.fit(X_train_feats, y_train)

# Set ridge weights
coef = ridge.coef_
intercept = ridge.intercept_
model.layers[-1].set_weights([coef.T, intercept])

# Evaluate
y_pred = model.predict(X_test)
y_pred_cls = np.argmax(y_pred, axis=1)
y_true_cls = np.argmax(y_test_cat, axis=1)

print(f"\nTest Accuracy: {accuracy_score(y_true_cls, y_pred_cls):.4f}")
print("\nClassification Report:")
print(classification_report(y_true_cls, y_pred_cls))
print("Confusion Matrix:\n", confusion_matrix(y_true_cls, y_pred_cls))
print(f"Cohen Kappa Score: {cohen_kappa_score(y_true_cls, y_pred_cls):.4f}")

# Plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Val loss')
plt.legend(); plt.title("Loss")

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train accuracy')
plt.plot(history.history['val_accuracy'], label='Val accuracy')
plt.legend(); plt.title("Accuracy")
plt.tight_layout()
plt.show()

print(f"⏱️ Total training + evaluation time: {time.time() - start_time:.1f} seconds")
```

```python
import numpy as np
import os, glob, scipy.io, time
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, cohen_kappa_score, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import RidgeClassifierCV

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (Input, SeparableConv1D, GlobalAveragePooling1D, BatchNormalization,
                                     Dropout, Multiply, Add, LayerNormalization, Dense, Lambda,
                                     Flatten, GaussianNoise, Concatenate)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import initializers

tf.keras.backend.clear_session()

# Ensure GPU
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"✅ GPU detected: {gpus[0]}")
    tf.config.experimental.set_memory_growth(gpus[0], True)
else:
    print("⚠️ No GPU found. Training might be slow.")

start_time = time.time()

# Load Data
folder_path = r"D:\abhishek_extracted\insomnia"
mat_files = sorted(glob.glob(os.path.join(folder_path, "*.mat")))[:39]

X_list, y_list = [], []
for file in mat_files:
    mat = scipy.io.loadmat(file)
    X_list.append(mat['ppg_signals'])
    y_list.append(mat['sleep_stages'].flatten())

X = np.vstack(X_list)
y = np.concatenate(y_list)

# Augmentation
def jitter(x, sigma=0.01): return x + np.random.normal(0, sigma, x.shape)
def scaling(x, sigma=0.1): return x * np.random.normal(1.0, sigma, x.shape)
def magnitude_warp(x, sigma=0.2):
    x = x.reshape(-1)
    warp = np.sin(np.linspace(0, np.pi, len(x))) * np.random.normal(1, sigma)
    return x * warp

np.random.seed(42)
augmented = []
target_distribution = np.bincount(y)
minority_classes = np.where(target_distribution < np.median(target_distribution))[0]

for signal, label in zip(X, y):
    if label in minority_classes:
        if np.random.rand() < 0.5:
            signal = jitter(signal)
        if np.random.rand() < 0.5:
            signal = magnitude_warp(signal)
        if np.random.rand() < 0.5:
            signal = scaling(signal.reshape(1, -1)).flatten()
    augmented.append(signal)

X = np.array(augmented)

# Normalize
scaler = StandardScaler()
X = scaler.fit_transform(X)

# SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Class weights
class_weights = compute_class_weight("balanced", classes=np.unique(y_res), y=y_res)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, stratify=y_res, random_state=42
)
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
num_classes = 5
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_test_cat = tf.keras.utils.to_categorical(y_test, num_classes=num_classes)

# Focal Loss
def focal_loss(gamma=1.5, alpha=0.25):
    def loss_fn(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.math.pow(1 - y_pred, gamma)
        return tf.reduce_mean(weight * cross_entropy)
    return loss_fn

# Fixed Random Layer for RVFL
class FixedRandomDense(tf.keras.layers.Layer):
    def __init__(self, units, activation='relu', **kwargs):
        super(FixedRandomDense, self).__init__(**kwargs)
        self.units = units
        self.activation = tf.keras.activations.get(activation)
        self.activation_str = activation

    def build(self, input_shape):
        self.w = self.add_weight(
            shape=(int(input_shape[-1]), self.units),
            initializer=initializers.RandomNormal(stddev=0.05),
            trainable=False,
            name='fixed_w'
        )
        self.b = self.add_weight(
            shape=(self.units,),
            initializer='zeros',
            trainable=False,
            name='fixed_b'
        )

    def call(self, inputs):
        return self.activation(tf.matmul(inputs, self.w) + self.b)

    def get_config(self):
        config = super(FixedRandomDense, self).get_config()
        config.update({
            "units": self.units,
            "activation": self.activation_str
        })
        return config

# Build CNN-RVFL Hybrid Model
def build_cnn_rvfl(input_shape, rvfl_units=128, final_units=num_classes):
    inputs = Input(shape=input_shape)
    x = GaussianNoise(0.05)(inputs)

    # CNN Feature Extractor
    x = SeparableConv1D(32, 7, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = SeparableConv1D(64, 5, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = SeparableConv1D(128, 3, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.2)(x)
    x = BatchNormalization()(x)

    # Residual connection
    res = tf.keras.layers.Conv1D(128, 1, strides=8, padding='same')(inputs)
    if x.shape[1] != res.shape[1]:
        res = tf.keras.layers.Resizing(x.shape[1], x.shape[2])(res)
    x = Add()([x, res])
    x = LayerNormalization()(x)

    # SE Attention
    se = GlobalAveragePooling1D()(x)
    se = Dense(64, activation='relu')(se)
    se = Dense(128, activation='sigmoid')(se)
    se = Lambda(lambda s: tf.expand_dims(s, 1))(se)
    x = Multiply()([x, se])
    x = GlobalAveragePooling1D()(x)

    # RVFL Branch
    x_flat = Flatten()(inputs)
    x_rvfl = FixedRandomDense(rvfl_units, activation='relu')(x_flat)

    # Merge
    concat = Concatenate()([x, x_rvfl])
    outputs = Dense(final_units, activation='softmax')(concat)

    model = Model(inputs, outputs)
    return model

# Build and compile model
model = build_cnn_rvfl((X_train.shape[1], 1), rvfl_units=128, final_units=num_classes)
model.compile(optimizer=Adam(1e-4), loss=focal_loss(), metrics=['accuracy'])
model.summary()

# Callbacks
callbacks = [
    EarlyStopping(patience=8, restore_best_weights=True),
    ModelCheckpoint("best_model.keras", save_best_only=True),
    ReduceLROnPlateau(patience=3, factor=0.5)
]

# Train model
history = model.fit(X_train, y_train_cat, validation_data=(X_test, y_test_cat),
                    epochs=100, batch_size=64, class_weight=class_weight_dict,
                    callbacks=callbacks, verbose=1)

# Extract features before classifier
intermediate_model = Model(inputs=model.input, outputs=model.layers[-2].output)
X_train_feats = intermediate_model.predict(X_train)

# Ridge classifier
ridge = RidgeClassifierCV(alphas=np.logspace(-6, 6, 13), scoring='accuracy', cv=5)
ridge.fit(X_train_feats, y_train)

# Set ridge weights
coef = ridge.coef_
intercept = ridge.intercept_
model.layers[-1].set_weights([coef.T, intercept])

# Evaluate
y_pred = model.predict(X_test)
y_pred_cls = np.argmax(y_pred, axis=1)
y_true_cls = np.argmax(y_test_cat, axis=1)

print(f"\nTest Accuracy: {accuracy_score(y_true_cls, y_pred_cls):.4f}")
print("\nClassification Report:")
print(classification_report(y_true_cls, y_pred_cls))
print("Confusion Matrix:\n", confusion_matrix(y_true_cls, y_pred_cls))
print(f"Cohen Kappa Score: {cohen_kappa_score(y_true_cls, y_pred_cls):.4f}")

# Plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Val loss')
plt.legend(); plt.title("Loss")

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train accuracy')
plt.plot(history.history['val_accuracy'], label='Val accuracy')
plt.legend(); plt.title("Accuracy")
plt.tight_layout()
plt.show()

print(f"⏱️ Total training + evaluation time: {time.time() - start_time:.1f} seconds")
```

```python
import numpy as np
import os, glob, scipy.io, time
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
# import tensorflow_probability as tfp
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, cohen_kappa_score, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import RidgeClassifierCV

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, SeparableConv1D, GlobalAveragePooling1D, BatchNormalization,
    Dropout, Multiply, Add, LayerNormalization, Dense, Lambda,
    Flatten, GaussianNoise, Concatenate, Conv1D, MaxPooling1D, UpSampling1D
)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import initializers

tf.keras.backend.clear_session()

# Ensure GPU
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"✅ GPU detected: {gpus[0]}")
    tf.config.experimental.set_memory_growth(gpus[0], True)
else:
    print("⚠️ No GPU found. Training might be slow.")

start_time = time.time()

class MixStyle(tf.keras.layers.Layer):
    def __init__(self, p=0.5, alpha=0.3, **kwargs):
        super(MixStyle, self).__init__(**kwargs)
        self.p = p
        self.alpha = alpha

    def call(self, x, training=None):
        if not training or tf.random.uniform([]) > self.p:
            return x

        B = tf.shape(x)[0]
        mu, var = tf.nn.moments(x, axes=[1], keepdims=True)
        sig = tf.sqrt(var + 1e-6)
        x_normed = (x - mu) / sig

        gamma1 = tf.random.gamma([B, 1, 1], self.alpha)
        gamma2 = tf.random.gamma([B, 1, 1], self.alpha)
        lam = gamma1 / (gamma1 + gamma2)

        perm = tf.random.shuffle(tf.range(B))
        mu2 = tf.gather(mu, perm)
        sig2 = tf.gather(sig, perm)

        mu_mix = mu * lam + mu2 * (1 - lam)
        sig_mix = sig * lam + sig2 * (1 - lam)

        return x_normed * sig_mix + mu_mix

    def get_config(self):
        config = super(MixStyle, self).get_config()
        config.update({
            "p": self.p,
            "alpha": self.alpha,
        })
        return config
    
# -------------------- Denoising Autoencoder --------------------
def build_denoising_autoencoder(input_shape):
    inputs = Input(shape=input_shape)
    x = Conv1D(16, 3, activation='relu', padding='same')(inputs)
    x = MaxPooling1D(2, padding='same')(x)
    x = Conv1D(8, 3, activation='relu', padding='same')(x)
    encoded = MaxPooling1D(2, padding='same')(x)
    x = Conv1D(8, 3, activation='relu', padding='same')(encoded)
    x = UpSampling1D(2)(x)
    x = Conv1D(16, 3, activation='relu', padding='same')(x)
    x = UpSampling1D(2)(x)
    decoded = Conv1D(1, 3, activation='sigmoid', padding='same')(x)
    return Model(inputs, decoded, name="DenoisingAutoencoder")

# Load Data
folder_path = r"D:\abhishek_extracted\insomnia"
mat_files = sorted(glob.glob(os.path.join(folder_path, "*.mat")))[:39]

X_list, y_list = [], []
for file in mat_files:
    mat = scipy.io.loadmat(file)
    X_list.append(mat['ppg_signals'])
    y_list.append(mat['sleep_stages'].flatten())

X = np.vstack(X_list)
y = np.concatenate(y_list)

# Augmentation
def jitter(x, sigma=0.01): return x + np.random.normal(0, sigma, x.shape)
def scaling(x, sigma=0.1): return x * np.random.normal(1.0, sigma, x.shape)
def magnitude_warp(x, sigma=0.2):
    x = x.reshape(-1)
    warp = np.sin(np.linspace(0, np.pi, len(x))) * np.random.normal(1, sigma)
    return x * warp

np.random.seed(42)
augmented = []
target_distribution = np.bincount(y)
minority_classes = np.where(target_distribution < np.median(target_distribution))[0]

for signal, label in zip(X, y):
    if label in minority_classes:
        if np.random.rand() < 0.5:
            signal = jitter(signal)
        if np.random.rand() < 0.5:
            signal = magnitude_warp(signal)
        if np.random.rand() < 0.5:
            signal = scaling(signal.reshape(1, -1)).flatten()
    augmented.append(signal)

X = np.array(augmented)

# Normalize
scaler = StandardScaler()
X = scaler.fit_transform(X)

# SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Class weights
class_weights = compute_class_weight("balanced", classes=np.unique(y_res), y=y_res)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, stratify=y_res, random_state=42
)
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
num_classes = 5
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_test_cat = tf.keras.utils.to_categorical(y_test, num_classes=num_classes)

# Focal Loss
def focal_loss(gamma=1.5, alpha=0.25):
    def loss_fn(y_true, y_pred):
        epsilon = tf.keras.backend.epsilon()
        y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)
        cross_entropy = -y_true * tf.math.log(y_pred)
        weight = alpha * tf.math.pow(1 - y_pred, gamma)
        return tf.reduce_mean(weight * cross_entropy)
    return loss_fn

# Fixed Random Layer for RVFL
class FixedRandomDense(tf.keras.layers.Layer):
    def __init__(self, units, activation='relu', **kwargs):
        super(FixedRandomDense, self).__init__(**kwargs)
        self.units = units
        self.activation = tf.keras.activations.get(activation)
        self.activation_str = activation

    def build(self, input_shape):
        self.w = self.add_weight(
            shape=(int(input_shape[-1]), self.units),
            initializer=initializers.RandomNormal(stddev=0.05),
            trainable=False,
            name='fixed_w'
        )
        self.b = self.add_weight(
            shape=(self.units,),
            initializer='zeros',
            trainable=False,
            name='fixed_b'
        )

    def call(self, inputs):
        return self.activation(tf.matmul(inputs, self.w) + self.b)

    def get_config(self):
        config = super(FixedRandomDense, self).get_config()
        config.update({
            "units": self.units,
            "activation": self.activation_str
        })
        return config

# Build CNN-RVFL Hybrid Model
def build_model(input_shape, rvfl_units=128, num_classes=5):
    input_raw = Input(shape=input_shape)

    # Denoising Autoencoder
    dae = build_denoising_autoencoder(input_shape)
    x_denoised = dae(input_raw)

    # CNN Feature Extractor
    x = SeparableConv1D(32, 7, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x_denoised)
    x = BatchNormalization()(x)
    x = MixStyle()(x)
    x = SeparableConv1D(64, 5, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = SeparableConv1D(128, 3, strides=2, padding='same', activation='relu', kernel_regularizer=l2(1e-4))(x)
    x = Dropout(0.2)(x)
    x = BatchNormalization()(x)

    # Residual
    res = Conv1D(128, 1, strides=8, padding='same')(x_denoised)
    x = Add()([x, res])
    x = LayerNormalization()(x)

    # Attention
    se = GlobalAveragePooling1D()(x)
    se = Dense(64, activation='relu')(se)
    se = Dense(128, activation='sigmoid')(se)
    se = Lambda(lambda s: tf.expand_dims(s, 1))(se)
    x = Multiply()([x, se])
    x = GlobalAveragePooling1D()(x)

    # RVFL Branch
    flat = Flatten()(x_denoised)
    rvfl_out = FixedRandomDense(rvfl_units, activation='relu')(flat)

    # Merge
    concat = Concatenate()([x, rvfl_out])
    outputs = Dense(num_classes, activation='softmax')(concat)

    return Model(input_raw, outputs)

# Build and compile model
model = build_model((X_train.shape[1], 1), rvfl_units=128, num_classes=num_classes)
model.compile(optimizer=Adam(1e-4), loss=focal_loss(), metrics=['accuracy'])
model.summary()

# Callbacks
callbacks = [
    EarlyStopping(patience=8, restore_best_weights=True),
    ModelCheckpoint("best_model.keras", save_best_only=True),
    ReduceLROnPlateau(patience=3, factor=0.5)
]

# Train model
history = model.fit(X_train, y_train_cat, validation_data=(X_test, y_test_cat),
                    epochs=100, batch_size=32, class_weight=class_weight_dict,
                    callbacks=callbacks, verbose=1)

# Extract features before classifier
intermediate_model = Model(inputs=model.input, outputs=model.layers[-2].output)
X_train_feats = intermediate_model.predict(X_train)

# Ridge classifier
ridge = RidgeClassifierCV(alphas=np.logspace(-6, 6, 13), scoring='accuracy', cv=5)
ridge.fit(X_train_feats, y_train)

# Set ridge weights
coef = ridge.coef_
intercept = ridge.intercept_
model.layers[-1].set_weights([coef.T, intercept])

# Evaluate
y_pred = model.predict(X_test)
y_pred_cls = np.argmax(y_pred, axis=1)
y_true_cls = np.argmax(y_test_cat, axis=1)

print(f"\nTest Accuracy: {accuracy_score(y_true_cls, y_pred_cls):.4f}")
print("\nClassification Report:")
print(classification_report(y_true_cls, y_pred_cls))
print("Confusion Matrix:\n", confusion_matrix(y_true_cls, y_pred_cls))
print(f"Cohen Kappa Score: {cohen_kappa_score(y_true_cls, y_pred_cls):.4f}")

# Plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Val loss')
plt.legend(); plt.title("Loss")

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train accuracy')
plt.plot(history.history['val_accuracy'], label='Val accuracy')
plt.legend(); plt.title("Accuracy")
plt.tight_layout()
plt.show()

print(f"⏱️ Total training + evaluation time: {time.time() - start_time:.1f} seconds")
```

```python
import numpy as np
import os, glob, scipy.io, time
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
# import tensorflow_probability as tfp
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix, cohen_kappa_score, accuracy_score
from sklearn.utils.class_weight import compute_class_weight
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import RidgeClassifierCV

from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, SeparableConv1D, GlobalAveragePooling1D, BatchNormalization,
    Dropout, Multiply, Add, LayerNormalization, Dense, Lambda,
    Flatten, GaussianNoise, Concatenate, Conv1D, MaxPooling1D, UpSampling1D
)
from tensorflow.keras.regularizers import l2
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras import initializers

tf.keras.backend.clear_session()

# Ensure GPU
gpus = tf.config.list_physical_devices('GPU')
if gpus:
    print(f"✅ GPU detected: {gpus[0]}")
    tf.config.experimental.set_memory_growth(gpus[0], True)
else:
    print("⚠️ No GPU found. Training might be slow.")

start_time = time.time()

# Load Data
folder_path = r"D:\abhishek_extracted\insomnia"
mat_files = sorted(glob.glob(os.path.join(folder_path, "*.mat")))[:39]

X_list, y_list = [], []
for file in mat_files:
    mat = scipy.io.loadmat(file)
    X_list.append(mat['ppg_signals'])
    y_list.append(mat['sleep_stages'].flatten())

X = np.vstack(X_list)
y = np.concatenate(y_list)

# Augmentation
def jitter(x, sigma=0.01): return x + np.random.normal(0, sigma, x.shape)
def scaling(x, sigma=0.1): return x * np.random.normal(1.0, sigma, x.shape)
def magnitude_warp(x, sigma=0.2):
    x = x.reshape(-1)
    warp = np.sin(np.linspace(0, np.pi, len(x))) * np.random.normal(1, sigma)
    return x * warp

np.random.seed(42)
augmented = []
target_distribution = np.bincount(y)
minority_classes = np.where(target_distribution < np.median(target_distribution))[0]

for signal, label in zip(X, y):
    if label in minority_classes:
        if np.random.rand() < 0.5:
            signal = jitter(signal)
        if np.random.rand() < 0.5:
            signal = magnitude_warp(signal)
        if np.random.rand() < 0.5:
            signal = scaling(signal.reshape(1, -1)).flatten()
    augmented.append(signal)

X = np.array(augmented)

# Normalize
scaler = StandardScaler()
X = scaler.fit_transform(X)

# SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X, y)

# Class weights
class_weights = compute_class_weight("balanced", classes=np.unique(y_res), y=y_res)
class_weight_dict = {i: class_weights[i] for i in range(len(class_weights))}

# Split
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, stratify=y_res, random_state=42
)
X_train = X_train[..., np.newaxis]
X_test = X_test[..., np.newaxis]
num_classes = 5
y_train_cat = tf.keras.utils.to_categorical(y_train, num_classes=num_classes)
y_test_cat = tf.keras.utils.to_categorical(y_test, num_classes=num_classes)


# Fixed Random Layer for RVFL
class FixedRandomDense(tf.keras.layers.Layer):
    def __init__(self, units, activation='relu', **kwargs):
        super(FixedRandomDense, self).__init__(**kwargs)
        self.units = units
        self.activation = tf.keras.activations.get(activation)
        self.activation_str = activation

    def build(self, input_shape):
        self.w = self.add_weight(
            shape=(int(input_shape[-1]), self.units),
            initializer=initializers.RandomNormal(stddev=0.05),
            trainable=False,
            name='fixed_w'
        )
        self.b = self.add_weight(
            shape=(self.units,),
            initializer='zeros',
            trainable=False,
            name='fixed_b'
        )

    def call(self, inputs):
        return self.activation(tf.matmul(inputs, self.w) + self.b)

    def get_config(self):
        config = super(FixedRandomDense, self).get_config()
        config.update({
            "units": self.units,
            "activation": self.activation_str
        })
        return config

# Build CNN-RVFL Hybrid Model
from tensorflow.keras.regularizers import l2

def build_model(input_shape, rvfl_units=256, num_classes=5):
    input_raw = Input(shape=input_shape)

    # CNN Feature Extractor
    x = SeparableConv1D(32, 7, strides=2, padding='same', activation='relu',
                        kernel_regularizer=l2(1e-4))(input_raw)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    x = SeparableConv1D(64, 5, strides=2, padding='same', activation='relu',
                        kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)

    x = SeparableConv1D(128, 3, strides=2, padding='same', activation='relu',
                        kernel_regularizer=l2(1e-4))(x)
    x = BatchNormalization()(x)
    x = Dropout(0.3)(x)

    cnn_feat = GlobalAveragePooling1D()(x)

    # RVFL Head
    flat = Flatten()(input_raw)
    rvfl_out = Dense(rvfl_units, activation='relu',
                     kernel_regularizer=l2(1e-4),
                     trainable=True)(flat)  # Allow training for better adaptation

    rvfl_out = Dropout(0.3)(rvfl_out)  # Dropout on RVFL branch

    # Merge
    concat = Concatenate()([cnn_feat, rvfl_out])
    output = Dense(num_classes, activation='softmax',
                   kernel_regularizer=l2(1e-4))(concat)

    return Model(inputs=input_raw, outputs=output)

# Build and compile model
model = build_model((X_train.shape[1], 1), rvfl_units=128, num_classes=num_classes)
model.compile(
    optimizer=Adam(1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
model.summary()

callbacks = [
    EarlyStopping(patience=10, restore_best_weights=True, monitor='val_loss'),
    ReduceLROnPlateau(patience=3, factor=0.5, monitor='val_loss'),
    ModelCheckpoint("best_model.keras", save_best_only=True, monitor='val_loss')
]

history = model.fit(
    X_train, y_train_cat,
    validation_data=(X_test, y_test_cat),
    epochs=100,
    batch_size=64,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# Extract features before classifier
intermediate_model = Model(inputs=model.input, outputs=model.layers[-2].output)
X_train_feats = intermediate_model.predict(X_train)

# Ridge classifier
ridge = RidgeClassifierCV(alphas=np.logspace(-6, 6, 13), scoring='accuracy', cv=5)
ridge.fit(X_train_feats, y_train)

# Set ridge weights
coef = ridge.coef_
intercept = ridge.intercept_
model.layers[-1].set_weights([coef.T, intercept])

# Evaluate
y_pred = model.predict(X_test)
y_pred_cls = np.argmax(y_pred, axis=1)
y_true_cls = np.argmax(y_test_cat, axis=1)

print(f"\nTest Accuracy: {accuracy_score(y_true_cls, y_pred_cls):.4f}")
print("\nClassification Report:")
print(classification_report(y_true_cls, y_pred_cls))
print("Confusion Matrix:\n", confusion_matrix(y_true_cls, y_pred_cls))
print(f"Cohen Kappa Score: {cohen_kappa_score(y_true_cls, y_pred_cls):.4f}")

# Plot
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Train loss')
plt.plot(history.history['val_loss'], label='Val loss')
plt.legend(); plt.title("Loss")

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Train accuracy')
plt.plot(history.history['val_accuracy'], label='Val accuracy')
plt.legend(); plt.title("Accuracy")
plt.tight_layout()
plt.show()

print(f"⏱️ Total training + evaluation time: {time.time() - start_time:.1f} seconds")
```
