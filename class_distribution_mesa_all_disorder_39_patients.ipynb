```python
import scipy.io
import numpy as np
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter, defaultdict

# ‚úÖ Define your 4 folders (adjust paths accordingly)
folder_paths = {
    "Normal": r"C:\Users\DST\Desktop\micro_AK\normal",
    "Apnea": r"D:\abhishek_extracted\apnea",
    "PLM": r"D:\abhishek_extracted\plm",
    "Insomnia": r"D:\abhishek_extracted\insomnia"
}

# ‚úÖ Store results
all_stage_counts = {}
combined_counter = Counter()

# ‚úÖ Loop through each group
for group_name, path in folder_paths.items():
    mat_files = sorted(glob.glob(os.path.join(path, "*.mat")))[:39]
    labels = []

    for file in mat_files:
        mat = scipy.io.loadmat(file)
        if 'sleep_stages' in mat:
            labels.extend(mat['sleep_stages'].flatten())

    stage_counts = Counter(labels)
    all_stage_counts[group_name] = stage_counts
    combined_counter.update(stage_counts)

# ‚úÖ Normalize keys for plotting
all_stages = sorted(set(k for counts in all_stage_counts.values() for k in counts.keys()))

# ‚úÖ Plot individual distributions
fig, axs = plt.subplots(2, 2, figsize=(14, 10))
axs = axs.flatten()

for i, (group, counts) in enumerate(all_stage_counts.items()):
    values = [counts.get(stage, 0) for stage in all_stages]
    sns.barplot(x=all_stages, y=values, palette="viridis", ax=axs[i])
    axs[i].set_title(f"{group} Sleep Stage Distribution")
    axs[i].set_xlabel("Sleep Stage")
    axs[i].set_ylabel("Count")
    axs[i].set_xticks(all_stages)

plt.tight_layout()
plt.show()

# ‚úÖ Plot combined distribution
plt.figure(figsize=(8, 5))
combined_values = [combined_counter.get(stage, 0) for stage in all_stages]
sns.barplot(x=all_stages, y=combined_values, palette="magma")
plt.title("Combined Sleep Stage Distribution Across All Folders")
plt.xlabel("Sleep Stage")
plt.ylabel("Count")
plt.xticks(all_stages)
plt.show()

# ‚úÖ Print distribution summaries
print("\nüìù Class Distribution Summary:")
for group, counts in all_stage_counts.items():
    print(f"\nüîπ {group} Dataset:")
    for stage in all_stages:
        print(f"  Stage {stage}: {counts.get(stage, 0)} samples")

print("\nüî∏ Combined Distribution:")
for stage in all_stages:
    print(f"  Stage {stage}: {combined_counter.get(stage, 0)} samples")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Base data with NaNs for Apnea
data = {
    'Disorder': ['Normal'] * 5 + ['Insomnia'] * 5 + ['PLM'] * 5 + ['Apnea'] * 5,
    'Stage': ['Wake', 'N1', 'N2', 'N3', 'REM'] * 4,
    'Before Aug.': [
        16574, 3726, 16437, 3718, 5430,
        17650, 3518, 15613, 2590, 4760,
        20959, 3768, 14564, 1784, 3782,
        18163, 4348, 16370, 2114, 4751
    ],
    'After Aug.': [
        26531, 9242, 26322, 9229, 13596,
        28324, 8673, 24876, 6449, 11885,
        33442, 9401, 23409, 4426, 9453,
        29093, 10847, 26103, 5229, 11873
    ],
    'After SMOTE': [
        26531]*5 + [28324]*5 + [33442]*5 + [29093]*5
}

# Convert to DataFrame
df = pd.DataFrame(data)
df['Label'] = df['Disorder'] + ' - ' + df['Stage']
df = df.set_index('Label')[['Before Aug.', 'After Aug.', 'After SMOTE']]

# Prepare annotated text with count + percentage
annot = df.copy()

# Function to format count + percentage
def format_cell(value, total):
    if pd.isna(value) or total == 0 or pd.isna(total):
        return ""
    percent = (value / total) * 100
    return f"{int(value)}\n({percent:.1f}%)"

# Compute per-disorder totals for each column
annotated = pd.DataFrame(index=df.index, columns=df.columns)

for col in df.columns:
    for disorder in ['Normal', 'Insomnia', 'PLM', 'Apnea']:
        idx = df.index.str.startswith(disorder)
        total = df.loc[idx, col].sum(skipna=True)
        for row in df.loc[idx].index:
            value = df.loc[row, col]
            annotated.loc[row, col] = format_cell(value, total)

# Plot the heatmap with annotated cells
plt.figure(figsize=(10, 7))
sns.heatmap(df, annot=annotated, fmt='', cmap='YlGnBu', cbar=True,
            linewidths=0.5, linecolor='gray', mask=df.isna(), annot_kws={"fontsize": 8})

# plt.title("Class Distributions (Count and %) Across Stages and Disorders", fontsize=13)
plt.ylabel("Disorder - Sleep Stage", fontsize=11)
plt.xlabel("Processing Stage", fontsize=11)
plt.xticks(fontsize=9)
plt.yticks(fontsize=8)
plt.tight_layout()
plt.savefig("class_distribution_heatmap_annotated.png", dpi=300)
plt.show()
```
